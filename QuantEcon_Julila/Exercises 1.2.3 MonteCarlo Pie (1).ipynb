Exercise 1. Factorial(n)

We're going to try to use a recursive solution, unlike what the question poses, but hey...

```julia
function factorial2(n)
    (n==1) ? 1 : n*factorial2(n-1)
end
```

```julia
x = rand(1000,1)
y = rand(1000,1)
sum(convert.(Int,(x.^2 .+ y.^2 .< 1)))
```

```julia
for n in 1:10
    println(factorial2(n))
end
```

Exercise 2. Binomial Random Variable $Y \sim Binom(n,p)$

```julia
function binomial_rv(n, p)
    sum(convert.(Int, rand(n,1).<p))
end
```

```julia
binomial_rv(10,.1)
```

```julia
using Distributions
using Plots
gr()

function AR_1(ts_length, α) # \alpha(Tab)
    x = Array{Float64}(ts_length)
    x[1] = 0
    for j = 2:ts_length
        x[j] = α*x[j-1] + rand(Normal())
    end
    return x
end    
plot(AR_1(200,0.8), label = "alpha=0.8")
plot!(AR_1(200,0), label = "alpha=0.0")
plot!(AR_1(200,.98), label = "alpha=0.98")
```
