```julia
function one_1(p)
    p ? 1 : 0
end
```

```julia
function f(k:: Integer, n:: Integer)
    return k/2^n
end
```

```julia
f(4,3)
one_1(5>=f(9-1,10) && 5<= f(9,10))
```

```julia
function d(n:: Integer, x:: Real)
    S = 0
    for k in (1:(n*2^n))
        S = S + f(k-1,n) * one_1((f(k-1,n)<= x < f(k,n))) 
    end
    S = S + n*(x>=n)
    return S
end
```

```julia
using PyPlot
x_grid = linspace(0, 10, 100)
n_vals = [1, 2, 3, 4, 5]

function_vals = []
labels = []

for n1 in n_vals
    push!(function_vals, [d(n1, x1) for x1 in x_grid])
    push!(labels, "$n1")
end

push!(function_vals, x_grid)
push!(labels, "identity function")

fig, ax = subplots()
for n in n_vals
    ax[:plot](x_grid,function_vals[n])
end
#plot(x_grid,function_vals[2])
```
